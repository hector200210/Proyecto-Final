# Metodo de la falsa posicion o regla falsa.
## Tarea 3
Sea $f(x) = 2x \cos(2x) - (x+1)^2$, con $a = -3$ y $b = -2$

```python
import numpy as np
import matplotlib.pyplot as plt

fx = lambda x: 2*x*np.cos(2*x)-(x+1)**2
error_rel = lambda vn, va: (np.abs(va - vn)/np.abs(vn))*100
xr = -2.1913
a = -3
b = -2

num_iteraciones = 5
#Completar
for i in range(num_iteraciones):
  aviejo = a
  bviejo = b
  xi = round(b - ((fx(b)*(a-b))/(fx(a)-fx(b))),4)
  fa = round(fx(a),4)
  fb = round(fx(b),4)
  fxi = round(fx(xi),4)
  er = round(error_rel(xr,xi),4) #round limita los decimales?

  if fa*fxi < 0:
    b = xi
  else:
    a = xi
  print('.....................................................................................................................')
  print(f'i = {i+1} | a = {aviejo} | b = {bviejo} | xi = {xi} | f(a) = {fa} | f(b) = {fb} | f(xi) = {fxi} | er = {er} %')

x = np.linspace(-5,5,100)
plt.plot(x,fx(x))
plt.plot(xi,fx(xi),'ro') #son las coordenadas.
plt.grid()
plt.show()
```
