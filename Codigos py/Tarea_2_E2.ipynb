# Tarea 2 - Ejercicio  2; _Metodo de Newton - Raphson_
$f(x) = 8xsen(x){e^{-x}}-1$, con valor inicial $(x_0) = 0.3$.

Observamos que $f'(x) = 8{e^{-x}}[sen(x) + xcos(x) - xsen(x)]$

```python
import numpy as np
import matplotlib.pyplot as plt

fx = lambda x: 8*x*np.sin(x)*np.exp(-x)-1
fxx = lambda x: 8*np.exp(-x)*(np.sin(x) + x*np.cos(x) - x*np.sin(x))
error_rel = lambda va, vn: (np.abs(va - vn)/np.abs(va))*100
xviejo = 0.3

num_iteraciones = 5

for i in range(num_iteraciones):
  xi = round(xviejo-(fx(xviejo))/fxx(xviejo),4)
  er = round(error_rel(xi,xviejo),4)
  xviejo = xi
  print('..........................................................')
  print(f'i = {i+1} | xi = {xi} | er = {er} %')

x = np.linspace(-5,5,100)
plt.plot(x,fx(x))
plt.plot(xi,fx(xi),'ro')
plt.grid()
plt.show()
```
