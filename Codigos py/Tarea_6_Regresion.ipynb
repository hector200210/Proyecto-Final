# Regresion lineal
## Parte 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5.04, 8.12, 10.64, 13.18, 16.20, 20.04])

# Cantidad de datos
n = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

a1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)

x_mean = np.mean(x)
y_mean = np.mean(y)
a0 = y_mean - a1 * x_mean

# Imprimir resultados
print(f"Pendiente (a1): {a1:.4f}")
print(f"Ordenada al origen (a0): {a0:.4f}")
print(f"Función de regresión: f(x) = {a1:.4f}x + {a0:.4f}")

# Graficar los puntos y la recta de regresión
x_pred = np.linspace(min(x), max(x), 100)
y_pred = a0 + a1 * x_pred

plt.plot(x, y, 'o', label='Datos originales')
plt.plot(x_pred, y_pred, 'r-', label='Recta de regresión')
plt.title('Regresión Lineal por Mínimos Cuadrados')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## Parte 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv')
X = df['Experience Years']
Y = df['Salary']

# Cálculos para la regresión lineal
sum_xiyi = sum(X*Y)
sum_xi2 = sum(X**2)
n = len(X)

a1 = (n*sum_xiyi - sum(X)*sum(Y)) / (n*sum_xi2 - sum(X)**2)
a0 = (sum(Y)/n) - a1 * (sum(X)/n)

# Datos para la gráfica de regresión
x_grafica = np.linspace(min(X), max(X), 100)
y_grafica = a0 + a1 * x_grafica

# Mostrar estimaciones
y_regresion = lambda x: a0 + a1 * x
datos_estimar = [15, 30, 50]
for dato in datos_estimar:
    print(f'Años {dato}: salario: {y_regresion(dato):.2f}')

print(f'La función de regresión es: f(x) = {a1:.2f}x + {a0:.2f}')

# Gráfica
plt.figure(figsize=(10, 6))  # Tamaño más grande
plt.plot(x_grafica, y_grafica, color='blue', linewidth=2.5, label='Línea de regresión')
plt.scatter(X, Y, color='darkorange', edgecolor='black', s=80, label='Datos reales')

# Personalización
plt.title('Experiencia vs Salario', fontsize=16)
plt.xlabel('Años de experiencia', fontsize=14)
plt.ylabel('Salario', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

plt.show()
```
